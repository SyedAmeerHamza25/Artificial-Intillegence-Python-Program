```python
input1=int(input("Enter Value 1 :"))
input2=int(input("Enter Value 2 :"))

operater=input("Enter value of operater")

if operater=="+":
    print(input1 + input2)
    
elif operater=="-":
    print(input1 - input2)
    
elif operater=="/":
    print(input1 / input2)
    
elif operater=="x":
    print(input1 * input2)
    
else:
    print("Please insert correct operater")
```
