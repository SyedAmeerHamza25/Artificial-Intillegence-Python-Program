```python
English =int(input("Enter English Marks:"))
Isl =int(input("Enter Isl Marks:"))
Maths =int(input("Enter Maths Marks:"))
Comp =int(input("Enter Comp Marks:"))
Urdu =int(input("Enter Urdu Marks:"))

obtain = English + Isl+ Maths+ Comp+ Urdu

perc = obtain / 500 * 100 
print ("Your percentage is :" ,perc)

if perc >=80 and perc <=100:
    print("A-One Grade")
elif perc >= 70 and perc <80:
    print("A Grade")
elif perc >= 60 and perc <70:
    print("B Grade")
elif perc >= 50 and perc <60:
    print("C Grade")
elif perc >= 40 and perc <50:
    print("D Grade")
elif perc >= 33 and perc <40:
    print("E Grade")
else:
    print("Fail")
    
```
